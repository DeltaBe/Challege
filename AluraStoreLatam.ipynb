# 2. Ventas por categoría

```python
import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
categoria = tienda.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False)
categoria2 = tienda2.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False)
categoria3 = tienda3.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False)
categoria4 = tienda4.groupby('Categoría del Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False)
print(f'La cantidad de productos vendidos por la categoria de la tienda 1 es:\n {categoria}')
print(f'La cantidad de productos vendidos por la categoria de la tienda 2 es:\n {categoria2}')
print(f'La cantidad de productos vendidos por la categoria de la tienda 3 es:\n {categoria3}')
print(f'La cantidad de productos vendidos por la categoria de la tienda 4 es:\n {categoria4}')
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
promedio = tienda['Calificación'].mean()
promedio2 = tienda2['Calificación'].mean()
promedio3 = tienda3['Calificación'].mean()
promedio4 = tienda4['Calificación'].mean()
print(f'La calificación promedio de la tienda 1 es: \n{promedio.round(1)}')
print(f'La calificación promedio de la tienda 2 es: \n{promedio2.round(1)}')
print(f'La calificación promedio de la tienda 3 es: \n{promedio3.round(1)}')
print(f'La calificación promedio de la tienda 4 es: \n{promedio4.round(1)}')
```

# 4. Productos más y menos vendidos

```python
import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

```python
producto_mas_vendido = tienda.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).head(1)
producto_mas_vendido2 = tienda2.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).head(1)
producto_mas_vendido3 = tienda3.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).head(1)
producto_mas_vendido4 = tienda4.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).head(1)
print(f'El producto mas vendido de la tienda 1 es:\n{producto_mas_vendido}')
print(f'El producto mas vendido de la tienda 2 es:\n{producto_mas_vendido2}')
print(f'El producto mas vendido de la tienda 3 es:\n{producto_mas_vendido3}')
print(f'El producto mas vendido de la tienda 4 es:\n{producto_mas_vendido4}')
print('/n')
producto_menos_vendido = tienda.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).tail(1)
producto_menos_vendido2 = tienda2.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).tail(1)
producto_menos_vendido3 = tienda3.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).tail(1)
producto_menos_vendido4 = tienda4.groupby('Producto')['Cantidad de cuotas'].sum().reset_index().sort_values(by='Cantidad de cuotas', ascending=False).tail(1)
print(f'El producto menos vendido de la tienda 1 es:\n{producto_menos_vendido}')
print(f'El producto menos vendido de la tienda 2 es:\n{producto_menos_vendido2}')
print(f'El producto menos vendido de la tienda 3 es:\n{producto_menos_vendido3}')
print(f'El producto menos vendido de la tienda 4 es:\n{producto_menos_vendido4}')
```
